```python
import os
import torch
from fetch_dataset import CoronaryCalciumDataset
```

```python
def main():
    # 1. Define paths (Update these if your folder names are different)
    raw_data_path = "3dino_dataset/cocacoronarycalciumandchestcts-2/Gated_release_final"
    save_dir = "preprocessed_coca"

    # Create the folder to hold our optimized tensors
    os.makedirs(save_dir, exist_ok=True)

    # 2. Load the original slow dataset
    print("Initializing raw dataset...")
    raw_dataset = CoronaryCalciumDataset(data_dir=raw_data_path)

    print(f"Starting preprocessing for {len(raw_dataset)} patients...")

    # 3. Loop through and save each patient as a pure PyTorch Tensor
    valid_count = 0
    for i in range(len(raw_dataset)):
        try:
            # This triggers the slow DICOM reading and XML parsing
            vol, mask = raw_dataset[i]
            
            # Save them directly to the hard drive
            torch.save(vol, os.path.join(save_dir, f"vol_{valid_count}.pt"))
            torch.save(mask, os.path.join(save_dir, f"mask_{valid_count}.pt"))
            
            valid_count += 1
            
            if valid_count % 10 == 0:
                print(f"Successfully processed {valid_count} patients...")
                
        except Exception as e:
            # If a patient folder is corrupted or empty, we safely skip it
            print(f"Skipping patient {i} due to error: {e}")

    print(f"\nPreprocessing complete! Saved {valid_count} valid patients to '{save_dir}'.")
```

```python
if __name__ == "__main__":
    main()
```
